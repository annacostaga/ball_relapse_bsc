# Check number of TFs in list of genes

```python
#import packages
import pandas as pd
import numpy as np
from gsynox import GSynoX
from gsynox.builder import Builder
import pandas as pd

g = GSynoX()
```

```python
#Load data
TFs = pd.read_csv("/home/spere5/ball_relapse_bsc/annotation/GRNs/GRNBoost2/allTFs_hg38.txt")
genes = pd.read_csv("/home/spere5/ball_relapse_bsc/annotation/GRNs/Filtering_GRN/Bcell_relevant_genes.csv", sep ="," )
genes = genes.iloc[: , 1:]
```

```python
# Collect relevant TFs
rel_TFs = genes[genes['Gene'].isin(TFs['TFs'])]
len(rel_TFs)
```

```python
# Find values in genes['Gene'] that are NOT in TFs['TFs']
genes_not_in_tfs = genes[~genes['Gene'].isin(TFs['TFs'])]
genes_not_in_tfs
```

```python
#Obtain synonyms for genes not found in TFs
genes_test = set(genes["Gene"])
synonyms = g.synonyms(genes_test)


```

```python
#Check if synonyms are in TF list
for syn_list in synonyms:
    if any(syn in TFs['TFs'].values for syn in syn_list):
        print(f"Matching Synonyms: {syn_list}")
```
